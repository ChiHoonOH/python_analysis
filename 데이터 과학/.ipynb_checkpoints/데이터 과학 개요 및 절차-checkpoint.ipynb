## 데이터 과학

**1. 정의**

- 데이터 과학은 데이터에서 어떤 의미나 통찰을 발견하고, 과학적인 방법으로 데이터를 탐색/분석/발견/통찰 등의 결과를 비즈니스 맥락에 활용 할 수 있는 S/W, system 구축을 포함

- 데이터 과학은 소프트웨어를 개발하면서 데이터를 분석하는 작업.(소프트웨어 개발자 관점)


**2. 핵심사항**

- 과학적인 방법으로 데이터를 탐색

> 질의 -> 가설수립 -> 수집/정제/구현/테스트 -> 결과 도출 및 평가
> 결과가 불만족 -> 가설로 이동
> 과학 실험과 동일한 방식

- 실제로 사용 가능한 소프트웨어 시스템 구축

> 구현 기술, 알고리즘 등이 비즈니스에 사용가능하여야 한다.

**3. history검토**

- 데이터 과학 분야의 출현은 통계학에서 시작한 것이 아니라, 
- 컴공이나 머신러닝분야에서 빅데이터를 다루면서 시작되었다.
- 향후 통계학, 데이터과학, 머신러닝, 딥러닝, 이런 분야들이 합쳐진 분야가 등장 할 것이라 예상

**4. 절차**

- 데이터 과학 로드맵
> 문제파악 -> 데이터분석 / 이해 => 특정값 추출 => 모델수립/분석 => 결과 정리 /보고서/발표 or 소프트웨어
> 모델수립/분석과정에서 만족스럽지 못하면 -> 문제파악으로 다시 return 
> 가설이나 질의가 실제로 의미가 있는지 끊임 없는 재질의가 필요.

- 결과물은 프로젝트의 성격에 따라 
> 문서,ppt,보고서 제공
> s/w(분석 코드, 실시간 분석 툴등등...)

- 문제 파악

> - 요구사항 분석
> - 상황에 맞는 정확한 질의 찾아낸다.
> - 어떤 기능이 필요한가?
> - 프로젝트의 끝은 어떤식인가?
> - 성능? 어떤 조건을 만족해야 하는지?
> - SOW : 외부에서 자문한다 -> 목표의 문서화, 작업 기술서 작성
> - PRD : 내부에서 프로젝트 -> 요구사항 문서

- 데이터분석 / 이해

 1. 데이터 이해
> - 데이터 셋의 크기
> - 제공된 데이터는 전체인지 OR 부분인지 
> - 데이터가 모집합인가, 부분집합인가
> - 데이터가 이상치인가?
> - 잡음(노이즈)이 있는 데이터인가?
> - 원본(raw data) or 가공데이터인지
> - 데이터 식별자가 존재하는가?
> - 데이터 머지와 연관성이 존재하는지?
> - 누락본이 존재하는지? 있다면 왜?
>
 2. 전처리.
> - 프로그램 작성, 데이터 로드 처리
> - 불필요한 데이터 삭제
> - 원하는 형태로 변환 -> 저장
> - 저장데이터가 작으면 csv, 적당한 볼륨 DB, 엄청 크다 -> hadoop
> - 문자열 데이터는 샘플링 확인 
> - 파일로 제공 되는 데이터는 파일 오픈 및 인덱스 등 확인  
> - 히스토그램, 산점도 등 데이터를 시각화해서 말이 되는 데이터인가? 체크
>
 3. 파일럿 분석 진행 -> 프로토 타입
> - 데이터를 가볍게 변환, 축소, 시각화
> - 머신러닝 분류기를 사용해도 좋음 
> - 데이터의 직관성을 키우기, 방향 검토
> - 시각화를 통해서 데이터의 성질과 가설도 세울 수 있음.


- 특정값 추출
> - 분석상 가장 중요한 단계
> - 해당분야에 대한 사전 지식 중요
> - 주식분야 분석(주식, 선물, 옵션, 파생상품 등등), html이해 
> - 머신러닝, 분류기의 통계 모델의 재료
> - 데이터를 가공하여서 표 테이블로 구성

- 모델수립/ 분석
> - 머신러닝을 사용 단계
> - 회귀모형(regression) , 군집화(clustering)
> - 데이터의 특성과 요구사항에 맞춰서 모델 적용
> - 신규모델을 만드는 경우는 드물게 있다.
> - 모델을 섞어서 사용하거나, 개선부분을 고려!! 


- 결과 정리/ 보고서/ 발표 or 소프트웨어
> - 슬라이드, 보고서 문서, 시각화 자료 등등 
> - 배치 코드 : 데이터 전체를 한번에 수행
> - 실시간 분석 코드 : 실시간 추적하여 분석 코드
> - 주석 포함 
> - 결과 검증 코드, 유닛 코드
> - 프로덕션 코드 : 고객사 배포 코드(인스톨러, 웹 등등)

- 반복
> - 데이터의 직관성이 생성될때까지 반복(파일럿 단계)
> - 산점도, 히스토그램
> - 불가능한 것인지 판단?
> - 전체 공정이나 일부 공정이 한번에 구동되도록 스크립트화



**5. 파이썬**

- 자료형의 해싱 체크 

> - 수정 불가 : 정수형, 문자형, 실수형, 튜플 등 -> 해싱 가능 자료형
> - 수정 가능 : 리스트, 딕셔너리, 기타 라이브러리 -> 해싱 불가 자료형

- 라이브러리 정리

> - pandas(팬더스, 판다스)  
> DataFrame, Series 형태로 데이터를 담아서 고속 처리, API, 이용하여 데이터 분석  
> R에 대응하는 모듈

> - numpy(넘파이)
> 수치 배열을 효율적 이용 . 수학 과학용 라이브러리

> - scikit-learn(사이킷 런)
> 넘파이 기반 머신러닝 라이브러리

> - matplotlib(맷플롯립)
> 데이터 시각화

> - scipy(사이파이)
> 넘파이 기반 빠른 수학 연산용

> - pyspark(파이스파크)
> 빅데이터 분야
