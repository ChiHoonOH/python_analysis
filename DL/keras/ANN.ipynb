### ANN(Artificial Neural Network)

- 생체 신경망 흉내낸 인공지능 
- 입력 - 은닉층 - 출력층
- ANN 인공신경망을 총칭하는 용어 
- 이중에 은닉층이 얕은 신경망 Shallow Neural Network : snn 으로 단일 은닉 계층을 구분 

### 개념 

- 초기 기술적 한계 때문에 은닉층 1개
- 빅데이터 이전 등장한 기술로, 데이터량은 적었음.
- 머신러닝쪽에 SVM 등이 더 좋은 성능을 내고 있어서 잘 안씀 
- 데이터가 많아지고, 비정형성의 데이터의 복잡도가 올라가면서 ANN을 각광받는 시기가 도래

- 1단계
> 입력계층을 통해서 외부로부터 들어오는 신호 벡터(x)
> 가중치 행렬 W(xh)을 곱해서 
> 은닉 계층으로 전송
> 반드시 입력 노드 중에 상수 1이 들어온다. 이값은 편항(bias) 보상에 사용한다. 

- 2단계
> 은닉 계층의 각 노드들   
> 입력신호 벡터에 활성화 함수 f(h)를 적용한 결과 벡터 h로 내보낸다.  
> 활성화함수   
> - 뉴런의 동작을 흉내, 비선형성을 보상하는 시그모이드, 하이퍼블릭 탄젠트 등 ... 

- 3단계  
> 은닉 계층의 결과물    
> 새로운 가중치 행렬 W(hy)를 곱한 후   
> 출력 계층으로 전송  

- 4단계
> 출력계층으로 들어온 신호 벡터에  
> 활성화 함수 f(y)를 적용  
> 결과 벡터 y를 신경망의 외부로 출력  
> 출력용 활성화 함수  
> - 주로 softmax 연산을 사용.  


### softmax 활성화 함수
- 입력받은 값을 출력으로 0~1사이의 값으로 모두 정규화를 수행하여 출력값들의 총합이 항상 1이 되게 만드는 연산을 수행하는 함수

- 일반적으로 NN 에서는 sigmoid function이 활성화 함수로 주로 사용된다.
- 그러나, 3가지 이상의 범주로 분류되는 다항분류의 경우는 softmax function을 많이 사용한다.
- 분류기 관련 응용에서는 softmax 가 좋은 성과를 낸다.

### ANN의 활용(2가지의 경우)

** 분류 **
- 입력 정보를 클래스별(손글씨의 경우 0~9)로 분류하는 방식
- 예
> MNIST 숫자를 0~9사이로 분류 => 분류 ANN
> 이런 분류에서는 하나의 클래스에 여러개의 값을 넣는 방식 보다, 각 값들을 이진화 처리하여 클래스를 분류하는 것이 더 성능이 좋다.

- 오차 역전파(error back propagation)
> - ANN을 이용하여 예측값을 추론하는 과정이 전방향 예측 경로   
> - ANN을 구성하는 가중치(W)의 학습 예측값의 목표값에 대해 오차를 역방향으로 되돌리면 진행. 오차역전파   
> - 오차 역전파는 오차를 줄이는 방법으로 경사하강법에 의해 유도.(경사하강법 때문에 등장 했다고 한다.)  
> - 경사하강법은 가중치에 대한 손실 함수를 미분하고, 그 미분값의 방향, 크기를 활용하여 가중치를 보상하는 방법  
> - 손실 함수는 가중치에 따라 오차가 얼마나 커지는지 작아지는지를 평가  
> - ANN은 손실함수로 교차 엔트로피 함수를 주로 사용. 이 함수를 적용하면 출력 노드의 결과를 확률값으로 변경되어야 하고 이 확률값은 
출력 노드를 softmax 연산을 통해서 구하게 된다.

- 오차 역전파 방법으로 학습 수행한다면 
> - 평균제곱근 오차 MSE(Mean square error)를 손실함수로 사용 
> - 최적화 방법으로 확률적 경사하강법, Adam, Adagrad, RMSprop 등등
> - 최적의 방법이란, 경험으로 알던가, 여러 방법의 조합으로 수행 후 결과를 선택하는 것이 최선이다.  

** 회귀 **

- 입력 정보로 다른값을 예측하는 방식  
- 예  
> 일주일간의 평균온도로 내일 평균온도 예측  
> 집값 에측, 시세 예측, ...  
> loss값 중요한 변수 -> loss가 작게 나오면 모델이 좋은 것
